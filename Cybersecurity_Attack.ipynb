##13.barplot
A bar plot visualize the categorical data using rectangular bars.

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
attack_plot=sns.countplot(x='Traffic_type',data=df,palette='viridis')
attack_plot.set_title('cybersecurity Attack Distribution')
attack_plot.set_ylabel('Number of Attacks')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
attack_plot=sns.countplot(x='Protocol',data=df,palette='viridis')
attack_plot.set_title('cybersecurity Attack Distribution')
attack_plot.set_ylabel('Number of Attacks')
plt.show()
```

##14.Heat Maps
Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps. In the below heat map we know that the Anomaly scores depends mainly on the Source port, Destination port, and Packet length

```python

sns.heatmap(df.corr(),vmin=-1,vmax=1)
plt.show
```

## 15. HISTOGRAMS
A histogram is a common EDA plot,which shows the distribution of a single variable

```python
df['Packet_length'].value_counts().head(20).plot(kind='hist',color='cyan',ec='black')
plt.title('Cybersecurity Attack')
plt.xlabel('Packet_length')
```

##16.piechart

```python
y=np.array([35,25,25,15])
mylabels=["Packet_length","Protocol","Packet_type","Source_port"]
plt.pie(y,labels=mylabels)
plt.show()
```

## 17.DISTPLOT
Displot is also known as the second Histogram because it is a slight improvement version of the Histogram.

```python
sns.distplot(df.skew(),color='blue',axlabel='Source_port')
```
